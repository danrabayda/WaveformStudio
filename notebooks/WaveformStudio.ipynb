```python
import os,IPython
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def normalize(s): #normalizes n-dimensional array-like s
    s = np.array(s)
    s = s - np.min(s)
    if np.max(s)!=0:
        s = s / np.max(s)
    return s
def kronecker_delta(i,j):
    return 1 if i==j else 0
def sample_gaussian(mean,std,length): #makes a gaussian in samples with given mean and std in terms of proportions
    return np.array([np.exp(-(((x/length)-mean)**2)/(2*(std**2))) for x in range(length)])
def sample_sigmoid(center,steepness,length): #makes a sigmoid in samples with center from 0 to 1 and steepness>=1, sign is 1 or -1 default is 1
    return np.array([1/(1+np.exp((int(center*length)-x)/(length/steepness))) for x in range(length)])
def sample_relu(center,length): #makes a relu centered at center
    return np.array([(x/length-center)/(1-center) if x/length>center else 0 for x in range(length)])

plt.plot([kronecker_delta(i,500) for i in range(1000)])
plt.plot(sample_relu(0.5,1000))
plt.plot(1-sample_sigmoid(0.5,10,1000))
plt.plot(sample_gaussian(0.5,0.1,1000))
plt.show()

plt.plot([np.cos(x*2*np.pi/500) for x in range(1000)])
plt.plot([np.sin(x*2*np.pi/500) for x in range(1000)])
plt.show()
```

```python
X=[x for x in range(1000)]
Y=[np.cos(x*2*np.pi/500) for x in range(1000)]
Z=[np.sin(x*2*np.pi/500) for x in range(1000)]

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(X,Y,Z)
plt.show()
```
